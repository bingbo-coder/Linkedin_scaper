```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
pd.set_option('display.max_rows', None)
df = pd.read_csv("database/linkedin_jobs_cleaned.csv")
df.drop(df.columns[[0,1]], axis=1, inplace= True)
df
```

```python
df.shape
```

```python
def position_semplifier(title):
    if "data scie" in title.lower(): 
        return "data scientist"
    elif "machine learning" in title.lower() or "ml" in title.lower():
        return "machine learning"
    elif "big data engineering" in title.lower():
        return "big data engineering"
    elif "software developer" in title.lower() or "sviluppatore di software" in title.lower() or "programmatore software" in title.lower() or "sviluppatore software" in title.lower():
        return "software developer"
    elif "software engineer" in title.lower():
        return "software engineer"
    elif "google cloud" in title.lower():
        return "google cloud"
    elif "big data" in title.lower():
        return "big data"
    elif "artificial intelligence" in title.lower():
        return "ai"
    elif "analista dati" in title.lower():
        return "data analyst"
    else:
        return "na"
    
def entry_position(title):
    if "junior" in title.lower():
        return "junior"
    elif "stage" in title.lower():
        return "stage"
    elif "internship" in title.lower():
        return "internship"
    else:
        return "na"
```

```python
df["job_simp"] = df["Position"].apply(position_semplifier)
df["entry_position"] = df["Position"].apply(entry_position)
job_simp = df["job_simp"].value_counts()
fig = px.bar(job_simp, x=job_simp.index, y=job_simp.values, title="Position Job"
            labels={
                     "index": "Position",
                     "y": "Count",
                 },)
fig.show()
```

```python
job = pd.crosstab(df['job_simp'],df['entry_position'])
job.drop("na", axis=1, inplace=True)
data = []
for x in job.columns:
    data.append(go.Bar(name=str(x), x=job.index, y=job[x]))

figure = go.Figure(data = data)
figure.update_layout(barmode = 'group')

figure.show()
```

```python
location = df["Location"].value_counts()
fig = px.bar(location, x=location.index[:10], y=location.values[:10], title="Job Location",
            labels={
                     "index": "City",
                     "y": "Count",
                     },)
fig.show()
```

```python
python = (df["python"].sum() / 400) * 100
excel = (df["excel"].sum() / 400) * 100
sql = (df["sql"].sum() / 400) * 100
power_bi = (df["power bi"].sum() / 400) * 100


skills_values = [python, excel, sql, power_bi]
skills = ["Python", "Excel", "SQL", "Power BI"]

print(skills_values)

fig = px.bar(x=skills, y=skills_values, title="Job Skills",
            labels={
                     "x": "Skill",
                     "y": "Percentage",
                     })
fig.show()
```

```python
def company_size(company):
    if company <= 200:
        return "small company"
    elif company <= 1000:
        return "medium company"
    else:
        return "big company "


df["company_size"] = df["N. employees"].apply(company_size)
company_size = df["company_size"].value_counts(normalize =True) *100

fig = px.bar(company_size, x=company_size.index, y=company_size.values, title="Company size",
            labels={
                     "index": "Company Size",
                     "y": "Percentage",
                     },)
fig.show()
```
